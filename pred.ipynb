```python
# Fix percentage columns
perc_cols = ['T1_Srv_Eff', 'T1_Rec_Pos', 'T1_Rec_Perf', 'T1_Att_Kill_Perc', 'T1_Att_Eff', 'T1_Att_Sum', 'T2_Srv_Eff', 'T2_Rec_Pos', 'T2_Rec_Perf', 'T2_Att_Kill_Perc', 'T2_Att_Eff', 'T2_Att_Sum']
for col in perc_cols:
    df[col] = pd.to_numeric(df[col].str.replace('%', ''))
```

```python
# Fix Date column to be datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y, %H:%M')
```

```python
# Fix values where decimals are represented with , instead of .
float_cols = ['T1_Srv_Err', 'T1_Blk_As', 'T2_Srv_Err', 'T2_Blk_As']
for col in float_cols:
    df[col] = pd.to_numeric(df[col].str.replace(',', '.'))
```

# Feature Engineering

```python
# Computing % of points gained from blocking
df['T1_Blk_Perc'] = df['T1_Blk_Sum']/df['T1_Sum']
df['T2_Blk_Perc'] = df['T2_Blk_Sum']/df['T2_Sum']
```

```python
for team in ['T1', 'T2']:
    # Computing % of points gained from blocking
    df[f'{team}_Blk_Perc'] = df[f'{team}_Blk_Sum']/df[f'{team}_Sum']

    # Computing % of aces from serves
    df[f'{team}_Srv_Ace_Rate'] = df[f'{team}_Srv_Ace'] / df[f'{team}_Srv_Sum']

    # Computing attack error rate
    df[f'{team}_Att_Err_Rate'] = df[f'{team}_Att_Err'] / df[f'{team}_Att_Sum']

    # Computing reception error rate
    df[f'{team}_Rec_Err_Rate'] = df[f'{team}_Rec_Err'] / df[f'{team}_Rec_Sum']

    # Computing overall score, average of serve, attack, and reception efficiency
    df[f'{team}_Overall_Score'] = (df[f'{team}_Srv_Eff'] + df[f'{team}_Att_Eff'] + df[f'{team}_Rec_Perf']) / 3
```

Additionally, my prior volleyball experience and training put a heavy emphasis on defense- as long as the ball doesn't hit the ground, there's always an opportunity to score. \
Because of this, I will also be creating a Defense Overall Score feature, which takes features regarding serve reception and blocking (unfortunately, the dataset lacks information regarding digs) \

```python
# Standardizing non-percent stats
min_val = min(df['T1_Blk_As'].min(), df['T2_Blk_As'].min())
max_val = max(df['T1_Blk_As'].max(), df['T2_Blk_As'].max())
df['T1_Blk_As_norm'] = (df['T1_Blk_As'] - min_val) / (max_val - min_val)

min_val = min(df['T1_Att_Blk'].min(), df['T2_Att_Blk'].min())
max_val = max(df['T1_Att_Blk'].max(), df['T2_Att_Blk'].max())
df['T1_Att_Blk_norm'] = (df['T1_Att_Blk'] - min_val) / (max_val - min_val)

df['T1_Def_Score'] = (
    df['T1_Rec_Perf'] * 0.4 # Good serve reception is key to the start of any point
    - df['T1_Rec_Err_Rate'] * 0.2 # Serve reception errors hurt
    + df['T1_Blk_Perc'] * 0.3 # Strong blocking is integral to strong defense
    + df['T1_Blk_As_norm'] * 0.1 # Counterattack-enabling blocks often convert to points
)

# Team 2 Defense Score
df['T2_Def_Score'] = (
    df['T2_Rec_Perf'] * 0.4
    - df['T2_Rec_Err'] * 0.2
    + df['T2_Blk_Sum'] * 0.3
    + df['T2_Blk_As'] * 0.1
)
```
